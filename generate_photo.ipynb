```python
def color_by_year(year):
    if year == 1774:
        col = "blue"
    elif year == 1786:
        col = "green"
    elif year == 1787:
        col = "cadetblue"
    elif year == 1788:
        col = "red"
    elif year == 1789:
        col = "orange"
    else:
        col = "purple"
    return col
```

```python
def color_by_tripnumber(tripno):
    if tripno == 0:
        col = "darkred"
    elif tripno == 1:
        col = "green"
    elif tripno == 2:
        col = "blue"
    elif tripno == 3:
        col = "red"
    elif tripno == 4:
        col = "orange"
    elif tripno == 5:
        col = "gray"
    elif tripno == 6:
        col = "pink"
    else:
        col = "purple"
    return col
```

```python
def add_markers(m,df):
    for i in range(len(df)):
        town = df.iloc[i]["Town"]
        season = df.iloc[i]["TripNumber"]
        lat = df.iloc[i]["Lat"]
        lon = df.iloc[i]["Lon"]
        year = df.iloc[i]["Year"]
        tripno = df.iloc[i]["TripNumber"]
        col = color_by_tripnumber(tripno)
        #col = color_by_year(year)
        if town == "London":
            col = "black"
        folium.Marker(
            location=[lat,lon],
            popup=town,
            icon=folium.Icon(color=col)
        ).add_to(m)
    
```

```python
m = folium.Map(location=[51.5,-0.13],zoom_start=5.25,width=750,height=1000)
add_markers(m,df)
```

```python
m2 = folium.Map(location=[51.5,-0.13],zoom_start=5.25,width=750,height=1000)
add_markers(m2,df2)
```
